```python
import pandas
import altair as alt
```

```python
df = pandas.read_csv("Enhanced_Major_Migration_Events_Dataset.csv")
df
```

```python
df["Start Year"]
```

```python
df['Start Year'] = df['Year'].str.extract(r'(\d{4})').astype(float)

selection = alt.selection_point(fields=['Start Year'], on="click")
color_condition = alt.condition(selection, "Event:N", alt.value("lightgray"))

points = alt.Chart(df).mark_circle(size=100).encode(
    x=alt.X(
        'Start Year:Q',
        scale=alt.Scale(domain=[1995, 2022]),
        axis=alt.Axis(format='d')
    ),
    y=alt.value(0),
    tooltip=[
        alt.Tooltip('Start Year:Q', title='Year'),
        alt.Tooltip('Event:N', title='Event'),
        alt.Tooltip('Description:N', title='Description'),
        alt.Tooltip('Migration Impact:N', title='Migration Impact')
    ],
    color=color_condition
).add_params(
    selection
).properties(
    width=800,
    height=100
)

points
```
